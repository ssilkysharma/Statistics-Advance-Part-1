2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define a discrete uniform distribution from a to b inclusive
a, b = 1, 5
rv = randint(a, b+1)  # b+1 because upper bound is exclusive in scipy

# Possible outcomes
x = np.arange(a, b+1)

# Compute PMF values
pmf = rv.pmf(x)

# Plot the PMF
plt.figure(figsize=(5, 4))
plt.bar(x, pmf, color='skyblue', edgecolor='black')
plt.title(f'PMF of Discrete Uniform Distribution U({a}, {b})')
plt.xlabel('x')
plt.ylabel('P(X = x)')
plt.xticks(x)
plt.ylim(0, pmf.max() + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
def bernoulli_pmf(x, p):
    """
    Compute the Bernoulli PMF:
      P(X = x) = p^x * (1-p)^(1-x)  for x in {0, 1}

    Args:
        x (int): Must be 0 or 1
        p (float): Probability of success, between 0 and 1

    Returns:
        float: The probability P(X = x)
    """
    if x not in (0, 1):
        return 0.0
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")
    # Formula from Bernoulli distribution
    return p**x * (1 - p)**(1 - x)

for x in [0, 1]:
    print(f"P(X={x}) = {bernoulli_pmf(x, p=0.6):.3f}")
```

4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Simulation parameters
n = 10     # number of trials per experiment
p = 0.5    # probability of success
size = 10000  # number of experiments to simulate

# Generate binomial random samples
data = np.random.binomial(n, p, size)

# Plot histogram of simulated data
plt.figure(figsize=(8, 5))
counts, bins, patches = plt.hist(data, bins=np.arange(-0.5, n + 1.5, 1),
                                 density=True, alpha=0.6, color='skyblue',
                                 edgecolor='black', label='Empirical')

# Overlay theoretical PMF for comparison
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
plt.vlines(x, 0, pmf, colors='red', linestyles='-', lw=2, label='Theoretical PMF')

plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(x)
plt.show()
```

 5.Create a Poisson distribution and visualize it using Python

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parameters
lambda_val = 4  # average number of events
k_max = 15      # maximum k-value to plot
sample_size = 1000

# 1. Compute PMF values
k_values = np.arange(0, k_max + 1)
pmf_values = poisson.pmf(k_values, mu=lambda_val)

# Plot PMF
plt.figure(figsize=(5,3))
plt.bar(k_values, pmf_values, color='violet', alpha=0.7, edgecolor='black')
plt.title(f'Poisson PMF (Œª = {lambda_val})')
plt.xlabel('Number of events (k)')
plt.ylabel('P(X = k)')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 2. Simulate random data and overlay histogram
data = np.random.poisson(lam=lambda_val, size=sample_size)

plt.figure(figsize=(5,3))
plt.hist(data, bins=np.arange(-0.5, max(data)+1), density=True,
         alpha=0.6, color='skyblue', edgecolor='black', label='Simulated')

# Overlay theoretical PMF
plt.plot(k_values, pmf_values, 'ro-', label='Theoretical PMF')
plt.title(f'Poisson Sample vs Theoretical (Œª = {lambda_val})')
plt.xlabel('Number of events')
plt.ylabel('Probability / Frequency')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for discrete uniform distribution
a, b = 1, 6

# Create the discrete uniform distribution: values in [a, b)
rv = randint(a, b+1)  # scipy defines randint as [a, b)

# Define the support (possible integer outcomes)
x = np.arange(a, b+1)

# Compute CDF at each value
cdf_values = rv.cdf(x)

# Plot the CDF: use a step function to represent jump behavior
plt.figure(figsize=(6, 4))
plt.step(x, cdf_values, where='post', linewidth=2, color='navy')
plt.scatter(x, cdf_values, color='navy')
plt.title(f'CDF of Discrete Uniform Distribution U({a}, {b})')
plt.xlabel('x')
plt.ylabel('F(x) = P(X ‚â§ x)')
plt.ylim(-0.05, 1.05)
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.xticks(x)
plt.show()
```

 10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_size = 10000
sample_size = 30
num_samples = 1000

# Generate a population from an exponential distribution
population = np.random.exponential(scale=1.0, size=population_size)

# Function to compute sample means
def compute_sample_means(population, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means

# Compute sample means
sample_means = compute_sample_means(population, sample_size, num_samples)

# Plot the population distribution
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, color='lightgreen', edgecolor='black', alpha=0.7, density=True)
plt.title(f'Sampling Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')

# Overlay a normal distribution for comparison
mean_of_sample_means = np.mean(sample_means)
std_dev_of_sample_means = np.std(sample_means)
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, norm.pdf(x, mean_of_sample_means, std_dev_of_sample_means), 'r-', lw=2, label='Normal Approximation')
plt.legend()

plt.tight_layout()
plt.show()

# Display summary statistics
print(f"Mean of sample means: {mean_of_sample_means}")
print(f"Standard deviation of sample means: {std_dev_of_sample_means}")
```

19. Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
import scipy.stats as stats
import math

# Given data
sample_mean = 980
population_mean = 1000
population_std_dev = 80
sample_size = 64
alpha = 0.05

# Step 1: Calculate the Z-Score
standard_error = population_std_dev / math.sqrt(sample_size)
z_score = (sample_mean - population_mean) / standard_error

# Step 2: Determine the critical Z-Value for a two-tailed test
z_critical = stats.norm.ppf(1 - alpha/2)

# Step 3: Make the decision
if abs(z_score) > z_critical:
    result = "Reject the null hypothesis: There is enough evidence that the average bulb life is less than 1000 hours."
else:
    result = "Fail to reject the null hypothesis: Not enough evidence that the average bulb life is less than 1000 hours."

# Output results
print(f"Calculated Z-score: {z_score:.3f}")
print(f"Critical Z-value: ¬±{z_critical:.3f}")
print(result)
```

 20.Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

# Sample data: Weights of 100 individuals
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)

# Define confidence level
confidence_level = 0.95

# Calculate standard error
standard_error = sample_std / np.sqrt(sample_size)

# Determine the critical value (z-score) for the confidence level
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_of_error = z_critical * standard_error

# Calculate confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Critical Z-value: {z_critical:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"{int(confidence_level * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

def analyze_distribution(data):
    """
    Calculate and print the skewness and kurtosis of the given dataset.
    Also, visualize the distribution with a histogram and a normal curve.
    """
    # Calculate skewness and kurtosis
    skewness_value = skew(data)
    kurtosis_value = kurtosis(data)

    # Print the results
    print(f"Skewness: {skewness_value:.4f}")
    print(f"Kurtosis: {kurtosis_value:.4f}")

    # Plotting the histogram
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)

    # Plotting the normal curve
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = 1/(np.sqrt(2 * np.pi * np.std(data)**2)) * np.exp(- 0.5 * ( x - np.mean(data))**2 / np.std(data)**2)
    plt.plot(x, p, 'k', linewidth=2)

    plt.title('Histogram and Normal Curve')
    plt.xlabel('Data Values')
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()

# Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Analyze the distribution
analyze_distribution(data)
```
